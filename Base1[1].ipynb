#### A. Regresion

1. Carga de datos

```python
import pandas as pd

df = pd.read_csv('dataset.arff.csv')
df
```

2. Limpiar base de datos

```python
df = df.drop('id', axis=1)  # Id no aporta nada relevante al DF
df
```

3. Asignar atributos y clase

```python
X1 = df.drop('price', axis=1)   # Atributos
y1 = df['price']                # Clase
```

Separar los datos de entrenamiento y testeo

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1,
    test_size = 0.20,
    train_size=0.80,
    random_state = 42
    )
```

4. Generar los modelos

Modelo Knn

```python
from sklearn.neighbors import KNeighborsRegressor

knn_rgr = KNeighborsRegressor(
    n_neighbors = 7,
    weights= 'distance'
)
```

Arbol de decision

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(
    criterion = "entropy",
    random_state = 42
)
```

Naive bayes

```python
from sklearn.naive_bayes import GaussianNB

Naive_bayes = GaussianNB(
    var_smoothing=1e-9  # Suavizado de la varianza
)
```

5. Entrenar los modelos

```python
knn_rgr.fit(X1_train,y1_train)
decision_tree.fit(X1_train, y1_train)
Naive_bayes.fit(X1_train, y1_train)
```

6. Predecir ejemplos

```python
knn_y1_pred = knn_rgr.predict(X1_test)
Dtree_y1_pred = decision_tree.predict(X1_test)
NB_y1_pred = Naive_bayes.predict(X1_test)
```

7. Evaluar modelos

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def evaluar_modelo(nombre, y_true, y_pred): # funcion para evaluar los modelos
    print(f"\nModelo: {nombre}")
    print("MSE:", mean_squared_error(y_true, y_pred))
    print("MAE:", mean_absolute_error(y_true, y_pred))
    print("R2 Score:", r2_score(y_true, y_pred))
```

```python
evaluar_modelo("KNN", y1_test, knn_y1_pred)
evaluar_modelo("Árbol de Decisión", y1_test, Dtree_y1_pred)
evaluar_modelo("Naive Bayes", y1_test, NB_y1_pred)  # Aunque el resultado sea bajo
```

8. Validación Cruzada 10-fold

```python
from sklearn.model_selection import cross_val_score

def validar_cruzada_regresion(nombre, modelo, X, y):
    print(f"\n Validación cruzada (10-fold) - {nombre}")
    scores = cross_val_score(modelo, X, y, cv=10, scoring='neg_mean_squared_error')
    mse_scores = -scores  # Convertir a positivos
    print(f"MSE promedio: {mse_scores.mean():.2f}")
    print(f"Desviación estándar: {mse_scores.std():.2f}")
```

```python
validar_cruzada_regresion("KNN", knn_rgr, X1, y1)
validar_cruzada_regresion("Árbol de Decisión", decision_tree, X1, y1)
validar_cruzada_regresion("Naive Bayes", Naive_bayes, X1, y1)
```
